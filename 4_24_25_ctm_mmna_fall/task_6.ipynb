```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import CubicSpline

# Define the function
f = lambda x: np.abs(x)

# Generate sample points
x = np.linspace(-2, 2, 10)
y = f(x)

# Perform cubic spline interpolation
spline = CubicSpline(x, y)

# Generate fine points for plotting
x_fine = np.linspace(-2, 2, 100)
y_fine = spline(x_fine)

# Plot the original function and the interpolation
plt.figure(figsize=(8, 5))
plt.plot(x_fine, f(x_fine), 'r--', label='|x| (True Function)')
plt.plot(x_fine, y_fine, 'b-', label='Cubic Spline Interpolation')
plt.scatter(x, y, color='black', label='Sample Points')
plt.legend()
plt.title('Cubic Spline Interpolation of |x|')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid()
plt.show()
```
