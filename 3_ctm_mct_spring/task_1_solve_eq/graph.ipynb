```python
import numpy as np
import matplotlib.pyplot as plt

l2_norm = np.array([0.148148, 0.0668086, 0.026176, 0.00971415, 0.0035165, 0.00125784, 0.000447296, 0.0001586])
l2_norm_log = np.log(l2_norm)

c_norm = np.array([0.296296, 0.181913, 0.105095, 0.0571047, 0.0298447, 0.0152663, 0.00772189, 0.00388348])
c_norm_log = np.log(c_norm)

x = np.array([2, 4, 8, 16, 32, 64, 128, 256])
x_log = np.log(x)

fig, ax1 = plt.subplots()

color = 'tab:blue'
ax1.set_xlabel('Шаг дискретизации')
ax1.set_ylabel('$l_2$ norm', color=color)
ax1.plot(x_log, l2_norm_log, color=color)
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()
color = 'tab:red'
ax2.set_ylabel('c norm', color=color)
ax2.plot(x_log, c_norm_log, color=color)
ax2.tick_params(axis='y', labelcolor=color)
```
